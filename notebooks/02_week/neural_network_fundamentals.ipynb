```python
import torch
import torch.nn as nn

# [EN] Define a linear layer that takes 4 input features and outputs 2 logits.
# [TR] 4 girdi özelliğini alıp 2 adet tahmin skoru (logit) üreten lineer katmanı tanımlıyoruz.
linear_layer = nn.Linear(in_features=4, out_features=2)

# [EN] Sample input data representing 1 instance with 4 numbers (e.g., room count, size, etc.)
# [TR] 4 sayıdan oluşan örnek bir girdi verisi (Örn: oda sayısı, m2 vb.)
sample_input = torch.tensor([[1.0, 2.0, 3.0, 4.0]])

# [EN] Pass the input through the layer to get initial raw predictions (logits).
# [TR] Veriyi katmandan geçirerek ilk ham tahmin skorlarını (logit) elde ediyoruz.
output_logits = linear_layer(sample_input)

print("[TR] Modelin Ürettiği Ham Tahmin Skorları (Logitler):")
print(output_logits)
```

```python
# [EN] Create the ReLU activation function.
# [TR] ReLU aktivasyon fonksiyonunu oluşturuyoruz.
relu = nn.ReLU()

# [EN] A list of negative and positive numbers.
# [TR] Negatif ve pozitif sayılardan oluşan örnek bir veri.
sample_data = torch.tensor([-2.0, -0.5, 0.0, 1.5, 3.0])

# [EN] Pass data through ReLU: Negatives become 0, positives stay the same.
# [TR] Veriyi ReLU'dan geçiriyoruz: Negatifler 0 olur, pozitifler aynı kalır.
activated_data = relu(sample_data)

print("[TR] ReLU Öncesi Veri :", sample_data)
print("[TR] ReLU Sonrası Veri:", activated_data)
```

```python
# [EN] Define a custom Neural Network by chaining layers together.
# [TR] Katmanları birbirine bağlayarak kendi Yapay Sinir Ağımızı oluşturuyoruz.
class SimpleClassifier(nn.Module):
    def __init__(self):
        super(SimpleClassifier, self).__init__()
        self.layer1 = nn.Linear(4, 8)   # Girdi katmanı (4 -> 8)
        self.relu = nn.ReLU()           # Esneklik katmanı
        self.layer2 = nn.Linear(8, 2)   # Çıktı katmanı (8 -> 2 sınıf)

    def forward(self, x):
        # [EN] Flow of data through the network
        # [TR] Verinin ağ içindeki akış sırası
        return self.layer2(self.relu(self.layer1(x)))

# [EN] Instantiate the model and the loss criterion.
# [TR] Modeli ve hatayı ölçecek kriteri (Loss fonksiyonunu) tanımlıyoruz.
model = SimpleClassifier()
criterion = nn.CrossEntropyLoss()

sample_x = torch.randn(1, 4)
true_label = torch.tensor([1]) # [TR] Gerçek doğru cevap 1. sınıf

# [EN] Make a prediction and measure how wrong the model is.
# [TR] Tahmin yapıp modelin ne kadar yanıldığını (Loss) hesaplıyoruz.
predicted_logits = model(sample_x)
loss = criterion(predicted_logits, true_label)

print("[TR] Modelin İlk Denemedeki Hata Skoru (Loss):", loss.item())
```

```python
import torch.optim as optim

# [EN] Optimizer (Adam) acts like a tutor, adjusting model weights to reduce loss.
# [TR] Optimizer (Adam), hatayı azaltmak için modelin katsayılarını ayarlayan bir öğretmen gibidir.
optimizer = optim.Adam(model.parameters(), lr=0.01)

# [EN] Generate 100 fake examples for practice.
# [TR] Pratik yapmak için 100 tane rastgele örnek veri üretiyoruz.
X_train = torch.randn(100, 4)
y_train = torch.randint(0, 2, (100,))

epochs = 20
print("--- [TR] Eğitim Başlıyor (Öğrenme Süreci) ---")
for epoch in range(epochs):
    model.train()

    # 1. Tahmin Et (Atış Yap)
    outputs = model(X_train)
    # 2. Hatayı Hesapla (Iskalama Miktarını Ölç)
    loss = criterion(outputs, y_train)

    # 3. Ders Çıkar ve Düzelt (Yayı Ayarla)
    optimizer.zero_grad() # Eski bilgileri temizle
    loss.backward()       # Hatanın kaynağını bul (Türev al)
    optimizer.step()      # Kendini güncelle (Öğren)

    if (epoch + 1) % 5 == 0:
        print(f"Tur [{epoch+1}/{epochs}] - Kalan Hata (Loss): {loss.item():.4f}")
```
