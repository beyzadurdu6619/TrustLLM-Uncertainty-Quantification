```python
import torch
import torch.nn.functional as F

# Bir Dil Modelinin ürettiği ham tahmin sayıları (Logitler) olsun:
# Örneğin model 3 kelime düşündü: ["Elma", "Araba", "Kitap"]
raw_logits = torch.tensor([2.0, 1.0, 0.1])

# Softmax uygulayarak bunu olasılık dağılımına çeviriyoruz
probabilities = F.softmax(raw_logits, dim=-1)

print("Elma Olma Olasılığı :", round(probabilities[0].item() * 100, 2), "%")
print("Araba Olma Olasılığı:", round(probabilities[1].item() * 100, 2), "%")
print("Kitap Olma Olasılığı:", round(probabilities[2].item() * 100, 2), "%")
```
