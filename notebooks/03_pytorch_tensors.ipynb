```python
import torch

# 1. CPU üzerinde iki matris oluştur
matrix_A = torch.rand(size=(3, 5))  # 3 satır, 5 sütun
matrix_B = torch.rand(size=(5, 2))  # 5 satır, 2 sütun

# 2. Matrisleri GPU'ya taşı
if torch.cuda.is_available():
    matrix_A = matrix_A.to("cuda")
    matrix_B = matrix_B.to("cuda")

# 3. GPU üzerinde Matris Çarpımı yap (Derin Öğrenmenin Temeli)
result = torch.matmul(matrix_A, matrix_B)

print("Sonuç Matris Boyutu:", result.shape)
print("Bulunduğu Cihaz:", result.device)
```
